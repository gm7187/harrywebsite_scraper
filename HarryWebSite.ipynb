```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
u='https://www.codewithharry.com/blog?number='
lst=[]
for i in range(1,8):
    lst.append(u+str(i))
```

```python
baseurl='https://www.codewithharry.com'
def HarryBlogs():
    blog_dict ={ 'Title':[],
                 'Description': [],
                 'Links': [],
               }

    for link in lst:
        r=requests.get(link)
        html=r.content
        soup=BeautifulSoup(html,'html.parser')

        blogs=soup.find_all('div',{'class':'blogs'})

        for b in blogs:
            blog_dict['Title'].append(b.h2.text)
            blog_dict['Description'].append(b.p.text)
            blog_dict['Links'].append(baseurl + b.h2.a['href'])
        
    return pd.DataFrame(blog_dict)

    
```

```python
HarryBlogs()
```

```python
def HarryVideos():
    r=requests.get('https://www.codewithharry.com/videos')
    html=r.content
    soup=BeautifulSoup(html,'html.parser')
    
    allDivs=soup.find_all('div',{'class':'col-md-6 col-lg-4 text-center'})
    
    videos_dict ={ 'Title':[],
                 'Description': [],
            'Images Links': [],
            'Links': [],
           }
    for t in allDivs:
        videos_dict['Title'].append(t.h2.text)
        videos_dict['Description'].append(t.p.text)
        videos_dict['Images Links'].append(t.img['src'])
        videos_dict['Links'].append(baseurl + t.a['href'])
    
    return pd.DataFrame(videos_dict)
    
    
    
```

```python
HarryVideos()
```
